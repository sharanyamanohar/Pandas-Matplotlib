# Working with pandas and Matplotlib

```python
import numpy as np
import pandas as pd
```

```python
#lets create 1 data frame to do operation and keep the dataframe constant by using seed
np.random.seed(20)
df=pd.DataFrame(np.random.rand(10,4),index=pd.date_range('2022-07-14',periods=10),columns=['a','b','c','d'])
df
```

```python
#lets add one columns
df['new']=df['a']+df['b']
df
```

```python
#lets add one more columns by using some operation
df['d_square_new']=df['d'].apply(lambda x:x*x)
df
```

```python
df['d_log_new']=df['d'].apply(lambda x:np.log(x))
df
```

```python
#from the last data set we got nan values on last column so how to filter the values
```

```python
def test(x):
    if x<0:
        return x
    else:
        return np.log(x)
```

```python
df['d_log_new_after_function']=df['d'].apply(test)
df
```

```python
#some function applying for columns a
def test(x):
    if x<0:
        return 0
    else:
        return 1
```

```python
df['a_test_function']=df['a'].apply(test)
df
```

```python
#sorting any column
df.sort_values(by='b')
```

```python
#one more columns is added with copy function
df['copy_a']=df['a']
df
```

```python
#check the datatypes
df.dtypes
```

```python
#taking a index to the columns
df['new_date']=df.index
df
```

```python
#coverting the date column into day,month,year
df['day']=df.index.day
df['month']=df.index.month
df['year']=df.index.year
df
```

```python
#find the day on the respective dates given
df['days']=df.index.day_name()
df
```
